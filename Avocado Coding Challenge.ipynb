```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the csv and assign it a name 
avocado_data = pd.read_csv('avocado.csv', header=0, parse_dates = True)
```

```python
#check its been read okay 
avocado_data.head() 
```

```python
#another check to make sure it's been read okay - checking number of rows is correct
len(avocado_data.index)
```

```python
#basic descriptive statistics 

avocado_data.describe(include = 'all')
```

```python
#what years are in the data 
avocado_data.year.unique()
```

```python
#what regions are in the data 
avocado_data.region.unique()
```

### Basic observations & conclusions: ###

avg price = $1.41

max price = $3.25

min price = $0.44 

The max price seems to be much higher than average. 
Wonder wheather the max price happened in isolation or is related to the less popular type of avocado? 

Date range 2015 - 2018 

54 regions with Albany being most frequent 

2 types of avocados with conventional being most popular  

```python
#creating average price against year graph
avocado_year = avocado_data.groupby('year').mean()

avocado_year.AveragePrice.plot(figsize=(15,5), title = 'Average Price (YoY)')
plt.ylabel('Average_Price')
```

This seems to suggest that 2017 saw a huge increase in the average price of avocados. 

But that it went back down to nearer it's pre-2017 price range by 2018. 

```python
#creating average price over time (mm yy) graph
avocado_date = avocado_data.groupby('Date').mean()

avocado_date.AveragePrice.plot(figsize=(15,5), title = 'Average Price (MoM)')
plt.ylabel('Average_Price')
```

Above shows peaks mostly in the second half of each year, with the average prices dipping at the start of each year, especially in February. 

Wonder if the peak in prices coinsides with peaks in volume of sales?

Are there more people buying avocados in Summer/Autumn hence an increase in prices? 

There was a huge spike in prices in Sept/Oct 2017, which aligns with the previous year by year breakdown which showed a spike in average price in 2017. 

```python
#volume of sales over time (mm yy) graph

#creating volume by date variable
volume_date = avocado_data['Total Volume'].groupby(avocado_data['Date']).sum()

#plotting volume against date, using volume/date variable 
volume_date.plot(color='red',figsize=(15,5), title = 'Volume (MoM)')
plt.ylabel('Total_volume')
```

Spikes in volumes at the start of each year, and decreases in the second half of the year. 

Spikes seem to align with when prices are at their lowest, and decreases are around increases in prices. Which makes sense - people are less likely to buy something when it's more expensive. 

```python
#avg price for each region bar chart 

#creating price x region variable
price_region = avocado_data['AveragePrice'].groupby(avocado_data['region']).mean()

#plot avg price for each region in a bar chart 
price_region.plot.bar(figsize=(15,5), title = 'Average Price x Region')
plt.ylabel('Average_Price')
```

```python
#total volume for each region bar chart 

#creating volume x region variable 
volume_region = avocado_data['Total Volume'].groupby(avocado_data['region']).sum()

#plotting volume x region 
volume_region.plot.bar(color='red',figsize = (15,5), title = 'Volume x Region')
```

The region data is a bit interesting/needs cleaning up. 

Need to exclude "total US" as this is skewing the result.

Further some "regions" are cities, some are states, some are larger territories/areas of the US. And this makes it slightly difficult to compare them accurately. 

Might be worth exluding (or creating a seperate chart for) "west", "midsouth", "southcentral", "northeast" as these don't appear to be cities/states but rather larger areas. 

```python
#avg price for each type of avocado bar chart 

#creating price x type variable
price_type = avocado_data['AveragePrice'].groupby(avocado_data['type']).mean()

#plot average price 
price_type.plot.bar(figsize = (15,5), title = 'Price vs Type')
```

As expected Organic avocados are more expensive than Conventional avocados

```python
#Avocado type prices for each city
avocado_conventional = avocado_city[avocado_city["type"].isin(['conventional'])]
avocado_organic = avocado_city[avocado_city["type"].isin(['organic'])]
 
price_conventional = avocado_conventional['AveragePrice'].groupby(avocado_conventional['region']).mean()
price_organic = avocado_organic['AveragePrice'].groupby(avocado_organic['region']).mean()

fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes

width = 0.4

price_conventional.plot(kind='bar', color='grey', width=width, position=1,figsize = (15,5), title = 'Price x Type - City')
price_organic.plot(kind='bar', color='green', width=width, position=0)

ax.grid(False)

plt.show()
```

```python
#Checking to see if excluding organic avocados makes a difference in the regional data 
volume_conventional = avocado_conventional['Total Volume'].groupby(avocado_conventional['region']).sum()

fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

volume_conventional.plot(kind='bar', color='red', ax=ax,figsize = (15,5), title = 'Volume x Value - City (Conventional Avocados)')
price_conventional.plot(kind='line', color='blue', ax=ax2)

ax.set_ylabel('Volume', color='red')
ax2.set_ylabel('Price', color='blue')

ax.grid(False)
ax2.grid(False)

plt.axhline(y=np.nanmean(price_conventional), color = 'blue', linestyle='dotted')

plt.show()
```

### Findings from effects of regional analysis ###

**There are clearly variations in both price and volume based on region.**

A lot of regions follow the trend that if the avocados are cheap they buy more. 

We saw this in high volume cities such as LosAngeles and Houston, where the average price was significantly lower than average. 

The cities where volume was highest: 
- LA 
- NY
- Houston
- Denver 
- San Fran
- Chicago


However, somewhere like New York has one of the highest average prices for avocados but also has one of the highest volumes. The same goes for Chicago and San Francisco. 

This definitely suggests that there's more at play than just a cost/volume relationship. 


What could be causing this: 
- Distance of region from avocado farms

- Demographic of residents in region: income, age, gender, job type, etc

- Weather in region 

- Geographic size of region 

- Population size of region 

Would be interesting to get this kind of data to compare against the regions.


**All regions buy conventional avocados much more than they buy organic - but this is not surprising as they are less expensive.**

However, even when we remove the organic avocados from the data set, the regional differences remain the same. And the cost/volume relationship still does not apply to all regions.
